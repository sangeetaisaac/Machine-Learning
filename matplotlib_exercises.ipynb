```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

## Bar Chart

```python
#Sample data: Product categories and Revenue
categories = ['Electronics', 'Clothing', 'Accessories', 'Home', 'Beauty']
revenue = [40000, 25000, 18000, 35000,22000]

#Create a bar chart
plt.bar(categories, revenue)

#Assign title
plt.title("Revenue by Category")

#Set the x and y labels
plt.xlabel("Categories")
plt.ylabel("Revenue")

#Display the bar chart
plt.show()
```

# Customize the bar chart

```python
#Customize the bar chart

#Set the color of the chart
plt.bar(categories, revenue, color='Red')

#Change the intervals of the y category from 5000 to 6000
plt.yticks(range(0, 50000, 6000))

#Assign title
plt.title("Revenue by Category")

#Set the x and y labels
plt.xlabel("Categories")
plt.ylabel("Revenue")

#Display the bar chart
plt.show()
```

## Add Revenue numbers to each bar

```python
categories = ['Electronics', 'Clothing', 'Accessories', 'Home', 'Beauty']
revenue = [40000, 25000, 18000, 35000,22000]

#Create a bar chart
plt.bar(categories, revenue)

#Assign title
plt.title("Revenue by Category")

#Set the x and y labels
plt.xlabel("Categories")
plt.ylabel("Revenue")

#add the revenue to each bar
for i, rev in enumerate(revenue):
  plt.text(i, rev, rev, ha = 'center', va = 'bottom', fontsize=10)

plt.show()
```
